```julia
function pico_generator(literals=3,variables=4,alternatives=3)
    variables_range = 1:variables
    list = (sort([sort(shuffle(variables_range)[1:literals]) for i in 1:alternatives]))
    negations = bitrand(alternatives,literals)
    nlist = [[(if (negations[i,j]) 0-1 else 1 end) for j in 1:literals] for i in 1:alternatives]
    formula = [nlist[i].*list[i] for i in 1:alternatives]
    return formula
end   
```

```julia
function sat_generator(literals=3,variables=4,alternatives=3)
    variables_range = ('A':'Z')[1:variables]
    list = (sort([sort(shuffle(variables_range)[1:literals]) for i in 1:alternatives]))
    negations = bitrand(alternatives,literals)
    nlist = [[(if negations[i,j] "~" else "" end) for j in 1:literals] for i in 1:alternatives]
    formula = join(["("*join(nlist[i].*list[i],"∨")*")" for i in 1:alternatives],"∧")
    list_p = [Int.(list[i]).-64 for i in 1:alternatives]
    nlist_p = [[(if (negations[i,j]) 0-1 else 1 end) for j in 1:literals] for i in 1:alternatives]
    formula_p = [nlist_p[i].*list_p[i] for i in 1:alternatives]
    iter_solution = PicoSAT.itersolve(formula_p)
    solution = [sign.(sign.(i).+1) for i in iter_solution]
    return formula, solution
end   
```
