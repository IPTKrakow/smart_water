```julia
using Plots
```

```julia
r = 1:1000

plot(r,r.^0,label="O(1)")
plot!(r,log2.(r),label="O(log(n))")
plot!(r,sqrt.(r),label="O(sqrt(n))")
plot!(r,r,label="O(n)")
plot!(r,r.^2,label="O(n^2)")
plot!(r,big(2) .^ r,label="O(2^n)")
plot!(r,factorial.(big.(r)),label="O(n!)")

plot!(aspect_ratio = :equal )
plot!(ylims=(first(r),last(r)))
plot!(xlims=(first(r),last(r)))
plot!(xscale=:log10,yscale=:log10)
plot!(legend=:outerright)
plot!(minorgrid=true)
plot!(xlabel="n")
plot!(ylabel="N")
```
